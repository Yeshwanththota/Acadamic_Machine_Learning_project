# Data preprocessing

```python
# From the data description, cabin feature is in deck/cabin number/side form.
# Here we are taking out deck and side and leaving cabin number which 
# doesn't effect the model
```

```python
data["Deck"] = data["Cabin"]
data["Side"] = data["Cabin"]

# GetDeck returns which deck the passenger is on
def GetDeck(value):
    if pd.isna(value) == False:
        return value[0]
    else:
        return value

# Get Side return 1 if the passenger is on port side, and 0 for starboard side.
def GetSide(value):
    if pd.isna(value):
        return value
    else:
        if value[-1] == "P":
            return 1
        else:
            return 0
```

```python
# applying above defined functions to get the values for deck and side in seperate columns
```

```python
data["Deck"] = data["Deck"].apply(GetDeck)
data["Side"] = data["Side"].apply(GetSide)
data.drop("Cabin",inplace=True,axis=1)
```

```python
data.head(5)
```

```python
# If any of the values for RoomService, FoodCourt, ShoppingMall, Spa, VRDeck features
# is null then the passenger didn't spend anything, so we can replace them with zero. 
```

```python
print(len(data[data['RoomService']==0]))
print(len(data[data['FoodCourt']==0]))
print(len(data[data['ShoppingMall']==0]))
print(len(data[data['Spa']==0]))
print(len(data[data['VRDeck']==0]))
```

```python
# Replacing all null values for amenities to zero
data["RoomService"].fillna(value=0,inplace=True)
data["FoodCourt"].fillna(value=0,inplace=True)
data["ShoppingMall"].fillna(value=0,inplace=True)
data["Spa"].fillna(value=0,inplace=True)
data["VRDeck"].fillna(value=0,inplace=True)
```

```python
# filling null values with False (Non VIP) status
data["VIP"].fillna(value=False,inplace=True)
```

```python
# filling the null values in age column with median
data["Age"] = data["Age"].fillna(data["Age"].median())
```

```python
# function to get passenger number from passenger ID
data["Pnum"] = data["PassengerId"]

def GetPnum(ID):
    return int(ID[5:])

data["Pnum"] = data["Pnum"].apply(GetPnum)

#Dropping name and Passenger ID columns
data.drop(["Name","PassengerId"],inplace=True,axis=1)
```

```python
# gettting the different deck names using value counts
data['Deck'].value_counts()
```

```python
# replacing all the boolean value columns into 0 and 1
```

```python
data["VIP"] = data["VIP"].map({True:1,False:0})
data["CryoSleep"] = data["CryoSleep"].map({True:1,False:0})
data["Transported"] = data["Transported"].map({True:1,False:0})
```

```python
# From the problem description, people who are in cryosleep do not spend 
# any money on amenities, so we can find the null values of cryosleep based on whether
# they spent money on any amenities or not
```

```python
index = 0
```

```python
# if cryosleep value is NaN and passenger didn't spent any money on amenities
# then they are in cryosleep else not
```

```python
def CryosleepNa(CryoSleep,RoomService,FoodCourt,ShoppingMall,Spa,VRDeck):
    global index
    if pd.isna(CryoSleep[index]) == False:
            ReturnValue = CryoSleep[index]
            index += 1
            return ReturnValue
        
    for x in [RoomService,FoodCourt,ShoppingMall,Spa,VRDeck]:
        if x[index] != 0:
            index += 1
            return 0.0
    index += 1
    return 1.0
```

```python
# applying the above defined function to fill the null values in cryosleep column
data["CryoSleep"] = data["CryoSleep"].apply(lambda x: CryosleepNa(data.CryoSleep,data.RoomService,
                                                                              data.FoodCourt,data.ShoppingMall,data.Spa,data.VRDeck))
index=0
```

```python
# for null values in side,Homeplanet,Destination and Deck
# we fill them according to their unique values in proportions
```

```python
import random
```

```python
print("Percentage of unique side values:")
print(data['Side'].value_counts(normalize=True))
print("\nPercentage of unique HomePlanet values:")
print(data['HomePlanet'].value_counts(normalize=True))
print("\nPercentage of unique Destination values:")
print(data['Destination'].value_counts(normalize=True))
print("\nPercentage of unique Deck values:")
print(data['Deck'].value_counts(normalize=True))
```

```python
# defining functions for filling null values for side,homepalnet,destination and deck features
# according to the above proportions
def FillingSideNaN(value):
    if pd.isna(value):
        return random.choices([0.0,1.0],weights=(51.6,48.4))[0]
    return value

def FillingHomeNaN(value):
    if pd.isna(value):
        return random.choices(["Earth","Europa","Mars"],weights=(54.1,25.1,20.8))[0]
    return value

def FillingDestinationNaN(value):
    if pd.isna(value):
        return random.choices(["TRAPPIST-1e","55 Cancri e","PSO J318.5-22"],weights=(69.5,21.1,9.4))[0]
    return value
def FillingDeckNaN(value):
    if pd.isna(value):
        return random.choices(["F","G","E","B","C","D","A","T"],weights=(32.9,30.1,10.3,9.2,8.8,5.6,3,0.05))[0]
    return value
```

```python
# applying all the functions defined above to fill the null values accordingly
data["Side"] = data["Side"].apply(FillingSideNaN)
data["HomePlanet"] = data["HomePlanet"].apply(FillingHomeNaN)
data["Destination"] = data["Destination"].apply(FillingDestinationNaN)
data["Deck"] = data["Deck"].apply(FillingDeckNaN)
```

```python
# converting all categorical columns to dummy variables using one hot encoding and adding back to dataframe
```

```python
homeplanet = pd.get_dummies(data["HomePlanet"],drop_first=True)
destination = pd.get_dummies(data["Destination"],drop_first=True)
deck = pd.get_dummies(data["Deck"],drop_first=True)


data = pd.concat([data.drop(["HomePlanet","Destination","Deck"],axis=1),homeplanet,destination,
                     deck],axis=1)
```
